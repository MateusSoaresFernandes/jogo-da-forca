```
# lista de palavras que o professor disponibilizou
lista_palavras = ["casa", "bola", "massa"]

print('''******************************
 JOGO DA FORCA 
******************************''')
# funçao '\n' da espaco na hora da execução
print("\n")

# while significa enquanto, ou seja, enquanto for verdadeiro
#ERRO: caso digite outro numero qualquer, vai prosseguir com o jogo, caso digite 0 vai encerrar
while True:
    inicio = int(input("Digite (1) para Inicar ou (0) para Sair: "))

    if inicio:
#dentro da variavel POS, len le a quantidade de caracteres em uma string. randint seleciona uma posicao aleatoria
#comecando do 0 e subtrai 1, ajuda pelo caelum
    pos=randint(0, len[lista_palavras]-1):
#variavel palavra esta chamando a lista que criamos no inicio do codigo
#[pos] esta chamando a condicao que criamos
#riscos esta chamando a variavel palavra junto com o contador len para saber a quantidade de "-" que vao ser adicionados
    palavra = lista_palavras (pos):
    riscos = [" _ "] * len(palavra)

        print("\n")
        print("Inicio do jogo!!!")

        erros = 0
#enquanto erros for menor ou igual a dois a variavel letra vai permitir que o jogador digite uma letra
        while erros <= 2 :
            letra = input("Digite uma letra: ")

#se a letra estiver na palavra, pos chama a palavra com find (localiza a primeira ocorrencia especiificada)
#range especifica o incremento ou o decremento, chama a variavel pos, len conta quantos caracteres tem na palavara e chama palavra

            if letra in palavra:
                pos = palavra.find(letra)
                for i in range(pos, len(palavra)):
#se a letra estiver na palavra dentro da lista, vai incluir a letra em riscos
                    if letra == palavra[i]:
                        riscos[i] = letra
#se nao vai contar um erro e vai pedir ao jogador para tentar novamente                        
            else:
                erros += 1
                print ("Ops... tente novamente!")

#print esta mostrando as letras acertadas dentro dos riscos, ajuda caaelum
            print('Palavra:\n', riscos)
#acertando ou nao, print esta mostrando a palavra escohida
        print(palavra)
#se o jogador nao quiser iniciar o jogo, print avisa que esta saindo e termina o loop
    else:
        print("Fim de jogo!!!")
        print("\n")
        print("Volte sempre!!!")
#break finaliza o loop
        break
```
